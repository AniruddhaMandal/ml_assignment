```python
import numpy as np
from itertools import combinations
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import rbf_kernel, pairwise_distances
from sklearn.cluster import KMeans
from joblib import Parallel, delayed
```

# Data Generation

```python
class SyntheticData():
    def __init__(self):
        pass
    def circle(self,n=1000,r=1,origin=(0,0),sigma=0.05):
        theta = 2*np.pi*np.random.uniform(0,1,n)
        noise = np.random.normal(scale=sigma,size=(n,2))
        return np.array([[r*np.cos(_theta)+origin[0]+e1,r*np.sin(_theta)+origin[1]+e2] for (_theta,(e1,e2)) in zip(theta,noise)])

    def moon(self,n=1000, r=1, origin=(0,0),sigma=0.05):
        start = 2*np.pi*np.random.uniform(0,1,1)
        end = start + np.pi
        theta = start + np.pi*np.random.uniform(0,1,n)
        noise = np.random.normal(scale=sigma,size=(n,2))
        return np.array([[r*np.cos(_theta)+origin[0]+e1,r*np.sin(_theta)+origin[1]+e2] for (_theta,(e1,e2)) in zip(theta,noise)])

    def blobs(self,n=1000, means=[[0,0],[1,1]], sigmas=[0.1,0.1]):
        data = []
        for m, s in zip(means,sigmas):
            for i in range(n):
                data.append(np.random.normal(m,(s,s),2))
        return np.array(data)
```

```python
syn_data = SyntheticData()

circle1 = syn_data.circle(n=1000)
circle2 = syn_data.circle(n=1000,r=1,origin=(np.sqrt(2),np.sqrt(2)))
circles = np.concatenate((circle1,circle2),axis=0)
c_circles = np.concatenate((np.zeros(1000),np.ones(1000)),axis=0)



moon1 = syn_data.moon(r=1, origin=(1,1))
moon2 = syn_data.moon()
moons = np.concatenate((moon1,moon2),axis=0)
c_moons = np.concatenate((np.zeros(1000),np.ones(1000)),axis=0)

blobs = syn_data.blobs(sigmas=[0.3,0.3])

fig, ax = plt.subplots(1,3,sharey=True,figsize=(12,4))
fig.suptitle("Synthetic Data")
ax[0].scatter(circles[:,0],circles[:,1],c=c_circles)
ax[0].set_title("Citcles")

ax[1].scatter(moons[:,0],moons[:,1],c=c_moons)
ax[1].set_title("Moon")

ax[2].scatter(blobs[:,0],blobs[:,1])
ax[2].set_title("Blobs")

plt.show()

print(blobs.shape)
```

# Similarity Matrix Calculation

```python
w_circles = rbf_kernel(circles,circles,gamma=0.5)
w_moons = rbf_kernel(moons,moons,gamma=0.5)
w_blobs = rbf_kernel(blobs,blobs,gamma=0.5)
```

# Normalized Laplacian

```python
def normalized_laplacian(M):
    degree_matrix = np.diag(M.sum(axis=1))
    _normalization_matrix = np.diag(np.sqrt(1/M.sum(axis=1)))
    _normalized_laplacian = np.identity(len(M))-_normalization_matrix@M@_normalization_matrix
    return _normalized_laplacian
```

```python
norm_lap_circles = normalized_laplacian(w_circles)
```

# Eigen Value Decomposition

```python
eig_value, eig_vect = np.linalg.eig(norm_lap_circles)
```

# Spectral Clustering and the Evaluation of Clusters

### Spectral Clustering Algorithm

```python
def SpectralClustering(x, n_clusters,gamma=0.1):
    A = rbf_kernel(x,x,gamma=gamma)
    D = np.diag(x.sum(axis=1))
    L = D - A
    eigen_values, eigen_vectors = np.linalg.eig(L)
    embd_data = eigen_vectors[:,:n_clusters]

    kmeans = KMeans(n_clusters).fit(embd_data)
    cmp = kmeans.labels_
    plt.scatter(x[:,0],x[:,1],c=cmp)
    plt.plot()
    return cmp
```

### Evalutaion with Silhouette Score

```python
def SilhouetteScore(X,labels, metric='euclidean', sample_size=None,n_jobs=1,**kwds):
    A = intra_cluster_distance(X,labels,metric=metric, n_jobs=n_jobs, **kwds)
    B = nearest_cluster_distance(X, labels,metric=metric, n_jobs=n_jobs, **kwds)

    score = (B - A)/ np.maximum(A,B)

    return score.sum()

def intra_cluster_distance(X,labels,metric,n_jobs=1,**kwds):
    intra_dist = np.zeros(labels.size,dtype=float)
    values = Parallel(n_jobs=n_jobs)(
        delayed(_intra_cluster_distance)(X[np.where(labels == label)[0]],metric=metric,**kwds)
        for label in np.unique(labels)
    )

    for label, values_ in zip(np.unique(labels),values):
        intra_dist[np.where(labels==label)[0]] = values_

    return intra_dist

def _intra_cluster_distance(X, metric, **kwds):
    distances = pairwise_distances(X, metric=metric, **kwds)
    return distances.sum(axis=1)/ (distances.shape[0]-1)

def nearest_cluster_distance(X, labels, metric, n_jobs, **kwds):
    inter_dist = np.empty(labels.size, dtype = float)
    inter_dist.fill(np.inf)

    unique_labels = np.unique(labels)

    values = Parallel(n_jobs=n_jobs)(delayed
    (_nearest_cluster_distance)(
        X[np.where(labels == label_a)[0]],
        X[np.where(labels == label_b)[0]],
        metric, **kwds)
        for label_a,label_b in combinations(unique_labels,2)
    )

    for (label_a, label_b), (values_a,values_b) in zip(combinations(unique_labels, 2), values):
            indices_a = np.where(labels == label_a)[0]
            inter_dist[indices_a] = np.minimum(values_a, inter_dist[indices_a])
            del indices_a
            indices_b = np.where(labels==label_b)[0]
            inter_dist[indices_b] = np.minimum(values_b, inter_dist[indices_b])
            del indices_b
    return inter_dist

def _nearest_cluster_distance(X_a,X_b, metric, **kwds):
    dist = pairwise_distances(X_a,X_b, metric=metric, **kwds)
    dist_a = dist.mean(axis=1)
    dist_b = dist.mean(axis=0)
    return dist_a, dist_b
```

### Circles Clustering

```python
cluster_labels = SpectralClustering(circles,2)
sil_score = SilhouetteScore(circles,cluster_labels,n_jobs=2)
print("Silhouette Scores: ", sil_score)
```

### Moons Clustering

```python
cluster_labels = SpectralClustering(moons,2)
sil_score = SilhouetteScore(moons,cluster_labels,n_jobs=2)
print("Silhouette Scores: ", sil_score)
```

# Blobs Clustering

```python
cluster_labels = SpectralClustering(blobs,2)
sil_score = SilhouetteScore(blobs,cluster_labels,n_jobs=2)
print("Silhouette Scores: ", sil_score)
```
